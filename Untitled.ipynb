# Leer datos

```python
import pandas as pd
```

```python
df = pd.read_csv("HeartFailureDataset.csv")
df.nunique()
```

```python
df = df.drop('id', axis = 1)
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

# Modelo KNN

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, learning_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (classification_report, ConfusionMatrixDisplay,
                             RocCurveDisplay, roc_auc_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, stratify=Y, test_size=0.2, random_state=66
)

neighbors_settings = range(1, 21)

pipe_grid = Pipeline([
    ("scaler", StandardScaler()),
    ("knn",    KNeighborsClassifier())
])

grid_search = GridSearchCV(
    pipe_grid,
    {"knn__n_neighbors": list(neighbors_settings)},
    cv=5, scoring="roc_auc", return_train_score=True
)
grid_search.fit(X_train, y_train)

best_k_grid  = grid_search.best_params_["knn__n_neighbors"]
cv_auc_grid  = grid_search.cv_results_["mean_test_score"]
cv_train_auc = grid_search.cv_results_["mean_train_score"]
gap          = cv_train_auc - cv_auc_grid

# Comparacion numerica
print(f"Mejor k  - GridSearchCV: {best_k_grid}  (CV AUC: {grid_search.best_score_:.6f})")
```

```python
# Predicciones del modelo optimo
y_pred  = grid_search.predict(X_test)
y_proba = grid_search.predict_proba(X_test)[:, 1]

# Learning curve
train_sizes, lc_train, lc_val = learning_curve(
    grid_search.best_estimator_,
    X_train, y_train,
    cv=5, scoring="roc_auc",
    train_sizes=np.linspace(0.1, 1.0, 10),
    shuffle=True, random_state=66
)
lc_train_mean = lc_train.mean(axis=1)
lc_train_std  = lc_train.std(axis=1)
lc_val_mean   = lc_val.mean(axis=1)
lc_val_std    = lc_val.std(axis=1)

ks = list(neighbors_settings)
matplotlib.rc_file_defaults()
```

```python
# Figura 1: Equivalencia Loop vs GridSearchCV
plt.plot(ks, cv_auc_grid, marker="^", linestyle="--", color="purple")
plt.axvline(best_k_grid, color="gray", linestyle=":")          # ← axvline, no avline
plt.title(f"Metodo B - GridSearchCV\nMejor k={best_k_grid}")
plt.xlabel("n_neighbors")
plt.legend(fontsize=9)
plt.tight_layout()
plt.show()

# Figura 2: Curva ROC + Matriz de confusion
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
RocCurveDisplay.from_estimator(
    grid_search.best_estimator_, X_test, y_test,
    ax=axes[0], color="darkorange", name=f"KNN k={best_k_grid}"
)
axes[0].plot([0, 1], [0, 1], "k--", label="Clasificador aleatorio")
axes[0].set_title(f"Curva ROC  (k={best_k_grid})")
axes[0].legend()

ConfusionMatrixDisplay.from_estimator(
    grid_search.best_estimator_, X_test, y_test,
    display_labels=["No cardio", "Cardio"],                    # ← etiquetas explícitas
    cmap="Blues", ax=axes[1]
)
axes[1].set_title(f"Matriz de confusion  (k={best_k_grid})")
plt.suptitle("Evaluacion del modelo optimo", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()

# Figura 3: Diagnostico de sobreajuste
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

ax = axes[0]
ax.plot(ks, cv_train_auc, label="Train AUC", marker="o", color="steelblue")
ax.plot(ks, cv_auc_grid,  label="Val AUC",   marker="s", color="darkorange")
ax.fill_between(ks, cv_auc_grid, cv_train_auc, alpha=0.15, color="red", label="Brecha")
ax.axvline(best_k_grid, color="gray", linestyle=":")
ax.set_title("Train AUC vs Val AUC por k")
ax.set_xlabel("n_neighbors")
ax.set_ylabel("ROC-AUC")
ax.legend(fontsize=8)

ax = axes[1]
ax.bar(ks, gap,
       color=["red" if g > 0.02 else "steelblue" for g in gap],
       alpha=0.8, edgecolor="white")
ax.axhline(0.02, color="red", linestyle="--", linewidth=1.2, label="Umbral 0.02")
ax.axvline(best_k_grid, color="gray", linestyle=":")
ax.set_title("Brecha Train - Val AUC")
ax.set_xlabel("n_neighbors")
ax.set_ylabel("Delta AUC")
ax.legend(fontsize=8)

ax = axes[2]
ax.plot(train_sizes, lc_train_mean, "o-", color="steelblue",  label="Train AUC")
ax.plot(train_sizes, lc_val_mean,   "s-", color="darkorange", label="Val AUC (CV-5)")
ax.fill_between(train_sizes,
                lc_train_mean - lc_train_std,
                lc_train_mean + lc_train_std, alpha=0.12, color="steelblue")
ax.fill_between(train_sizes,
                lc_val_mean - lc_val_std,
                lc_val_mean + lc_val_std,     alpha=0.12, color="darkorange")
ax.set_title(f"Learning curve  (k={best_k_grid})")
ax.set_xlabel("Tamano del conjunto de entrenamiento")
ax.set_ylabel("ROC-AUC")
ax.legend(fontsize=8)

plt.suptitle("Diagnostico de sobreajuste", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()

# Resumen final
print(f"k optimo        : {best_k_grid}")
print(f"CV Train AUC    : {cv_train_auc[best_k_grid-1]:.4f}")
print(f"CV Val AUC      : {cv_auc_grid[best_k_grid-1]:.4f}")
print(f"Test AUC final  : {roc_auc_score(y_test, y_proba):.4f}\n")
print(classification_report(y_test, y_pred, target_names=["No cardio", "Cardio"]))  # ← mismo fix
```

```python
import joblib
import os

# ── 1. Save the full pipeline (scaler + KNN) ──────────────────────────────────
os.makedirs("model", exist_ok=True)

joblib.dump(grid_search.best_estimator_, "model/knn_cardio_pipeline.pkl")
print("✅  Pipeline saved → model/knn_cardio_pipeline.pkl")

# ── 2. Save supporting metadata ───────────────────────────────────────────────
import json

metadata = {
    "best_k"        : int(best_k_grid),
    "cv_auc"        : float(grid_search.best_score_),
    "test_auc"      : float(roc_auc_score(y_test, y_proba)),
    "features"      : list(X.columns),
    "target_name"   : "cardio",
    "class_labels"  : ["No cardio", "Cardio"],
    "random_state"  : 66,
    "test_size"     : 0.2,
}

with open("model/metadata.json", "w") as f:
    json.dump(metadata, f, indent=2)
print("✅  Metadata saved  → model/metadata.json")
```

# Bayesian

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
vars_num = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']
vars_cat = ['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']
preprocessor = ColumnTransformer(
    transformers=[
        ('num',
         Pipeline([
             ('scaler',  StandardScaler())
         ]),
         vars_num),

        ('cat',
         Pipeline([
             ('encoder', OneHotEncoder(
                 drop='first',
                 handle_unknown='ignore',
                 sparse_output=False
             ))
         ]),
         vars_cat)
    ],
    remainder='drop'
)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("classifier", GaussianNB())
])
grid_search = GridSearchCV(pipeline, {}, cv=5, scoring='roc_auc')
grid_search.fit(X_train, y_train)

y_pred_hot  = grid_search.predict(X_test)
y_proba_hot = grid_search.predict_proba(X_test)[:, 1]

acc = accuracy_score(y_test, y_pred_hot)
auc = roc_auc_score(y_test, y_proba_hot)

print(f"Mejor precisión en validación cruzada: {grid_search.best_score_:.4f}")
print(f"Accuracy en test: {acc:.4f}")
print(f"AUC en test:      {auc:.4f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred_hot))
```

```python
import joblib
import os

os.makedirs("bayesian", exist_ok=True)
joblib.dump(grid_search, "bayesian/naive_bayes_gridsearch.pkl")
print("✅ Model saved → bayesian/naive_bayes_gridsearch.pkl")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.metrics import roc_auc_score
import time
```

```python
param_grid = {'classifier__n_estimators': [10, 25, 50, 75, 100], 'classifier__max_depth': [5, 10, 15]}

pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])

# Entrenamiento con GridSearchCV
start_train = time.time()
grid = GridSearchCV(pipeline, param_grid=param_grid, scoring='roc_auc', cv=3, n_jobs=-1)
grid.fit(X_train, y_train)
elapsed_train = time.time() - start_train

# Predicción
y_pred = grid.predict(X_test)
y_prob = grid.predict_proba(X_test)[:, 1]  

# Métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)
cm = confusion_matrix(y_test, y_pred)

print("\nGridSearchCV (scikit-learn) — RandomForestClassifier")
print("Mejores parámetros:", grid.best_params_)
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")
print(f"ROC AUC:   {auc:.4f}")
print("\nMatriz de confusión:")
print(cm)
print(f"\nTiempo de entrenamiento: {elapsed_train:.2f} segundos")
```

```python
import joblib
import os

os.makedirs("Random Forest", exist_ok=True)
joblib.dump(grid, "Random Forest/random_forest_gridsearch.pkl")
print("✅ Model saved → Random Forest/random_forest_gridsearch.pkl")
```
